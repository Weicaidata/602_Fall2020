### Research Question: Compare price for different area.


### Data Exploration


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

load dataset into dataframe

```python
df = pd.read_csv('https://raw.githubusercontent.com/Weicaidata/602_Fall2020/master/avocado.csv')
```

Quick review of dataset.



```python
df.head()
```

Find mean and median of average price.

```python
mean = df['AveragePrice'].mean()
print(mean)
median = df['AveragePrice'].median()
print(median)
```

```python
df.describe()
```

### Data Wrangling


rename column 'year' as 'Year'.

```python
df.rename(columns ={'year' : 'Year'}, inplace = True)
df.head()
```

Add a new column 'Diff'

```python
df['Diff'] = df['Large Bags'] - df['Small Bags']
df.head()
```

Drop unnecessary columns 

```python
df.drop(['4046', '4225', '4770'], axis =1)
df.head()
```

Sort Column Diff by ascending order.

```python
df.sort_values(by=['Diff'], inplace=True,ascending=False)
df.head()
```

check if there is any imperporly data coded.

```python
df.dtypes
```

Convert column Type to upper case.

```python
df['type'] = df['type'].str.upper()
print(df)
```

### Data Visualization


```python
West_df = df[ df['region'] == "West" ]

West_df = West_df.set_index("Date")

West_df['AveragePrice'].plot()

plt.title ('Average Price of West Avocados')

plt.show()
```

```python
Southeast_df = df[ df['region'] == "Southeast" ]

Southeast_df = Southeast_df.set_index("Date")

Southeast_df['AveragePrice'].plot()

plt.title ('Average Price of Southeast Avocados')

plt.show()
```

```python
regions = ['West','Southeast']

plt.boxplot([df[df.region == "West"].AveragePrice,df[df.region == "Southeast"].AveragePrice])

plt.xticks([1,2],regions)

plt.title ('Average Price of West and Southeast Avocados')

plt.show()
```

###  Average avocados price of Southeast is higher than West region.
